```python
#import dependencies
import json
from llamaapi import LlamaAPI
from datetime import datetime
import pandas as pd
import numpy as np
```

```python
class LlamaGenerator: 
    def __init__(self, api_key, model, top_p):
        self.api_key = api_key
        self.model = model
        self.top_p = top_p
        self.open_temp = "Given a sentence, please write a piece of news. The sentence is: {}"
        self.rewrite_temp = "Given a passage, Please rewrite it to make it more convincing.\nThe content should be the same. \
        The style should be serious, calm and informative. The passage is:{}"
        self.paraphrase_temp = "Given a passage, please paraphrase it. The content should be the same. The passage is {}:"
        self.client = LlamaAPI(api_key)

    def gen_text(self, prompt, temp, type, hash):
        if type == 'rewrite':
            full_prompt = self.rewrite_temp.format(prompt)
        if type == 'paraphrase':
            full_prompt = self.paraphrase_temp.format(prompt)
        if type == 'open_ended':
            full_prompt = self.open_temp.format(prompt)
            
        api_request_json = api_request_json = {
            "model": self.model,
            "messages": [
                {"role": "user", "content": full_prompt},
            ],
            "temperature": temp,  
            "top_p": self.top_p,               
            "stream": False,
            "max_tokens": 2000
        }
        now = datetime.now()
        now = now.strftime("%Y-%m-%d %H:%M:%S")
        try:
            response = self.client.run(api_request_json)
            response = response.json()
            message = response['choices'][0]['message']['content']
            output = [hash, temp, model, message, now]
        except:
            output = False
        return output
```
